고등어 도매 가격의 경우

```python
mack_dome =pd.read_excel('고등어도매_중품_10kg.xlsx') #엑셀파일을 받아와, csv파일로 변환
mack_dome.to_csv('고등어도매_10kg')
mack_dome = pd.read_csv('고등어도매_10kg')
mack_dome.head()
```

```python
mack_dome = mack_dome.transpose() #행과 열 뒤집기
mack_dome.reset_index(level=0, inplace=True) #인덱스가 된 일자를 열로 들여옴
mack_dome = mack_dome.drop(0, axis=0) #불필요한 첫번째 행 삭제
mack_dome.columns = ['일자', '평균','평년', '서울'] #컬럼 이름을 변경
mack_dome = mack_dome.drop(1, axis=0) #불필요한 두번쨰 행 삭제
mack_dome = mack_dome.drop(['평년','서울'], axis=1) #평년, 서울 열 삭제
mack_dome.reset_index(drop=True, inplace=True) #인덱스 초기화
```

```python
mack_dome
#위 전처리코드는 한 런타임에 복수로 실행하면 오류가 나니 다시 실행시키실 경우 엑셀파일부터 받아 런타임하시면 됩니다.
```

무 도매 가격의 경우

```python
radish_h_dome =pd.read_excel('무도매_상품_20kg.xlsx')
radish_h_dome.to_csv('무도매_상품_20kg')
radish_h_dome = pd.read_csv('무도매_상품_20kg')
radish_h_dome.head()
```

```python
radish_h_dome = radish_h_dome.drop([1,2,3,4,5], axis=0) #불필요한 첫번째 행 삭제
radish_h_dome = radish_h_dome.transpose() #행과 열 뒤집기
radish_h_dome.reset_index(level=0, inplace=True) #인덱스가 된 일자를 열로 들여옴
radish_h_dome.columns = ['일자', '평균'] #컬럼 이름을 변경
radish_h_dome = radish_h_dome.drop([0,1,2], axis=0) #쓸모없는 0, 1, 2 인덱스의 행 삭제
radish_h_dome.reset_index(drop=True, inplace=True) #인덱스 초기화
```
